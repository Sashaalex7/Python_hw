```python
HOUSING_PATH = "data/housing.csv"
```

```python
import pandas as pd

housing = pd.read_csv(HOUSING_PATH)
housing.head()
```

1. Масштабирование:

```python
#а. Нормализовать признак - median_income
```

```python
from sklearn import preprocessing
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
```

```python
housing_m_in = housing[["median_income"]]
```

```python
housing_m_in_scaled = min_max_scaler.fit_transform(housing_m_in)
```

```python
housing_m_in_scaled
```

```python
#б. Привести признак - population к масштабу от 0 до 100
```

```python
min_max_scaler_2 = preprocessing.MinMaxScaler(feature_range=(0,100))
```

```python
housing_population = housing[["population"]]
```

```python
housing_population_scaled = min_max_scaler_2.fit_transform(housing_population)
```

```python
housing_population_scaled
```

2. Заполнение пропусков:

```python
#а. Заполнить пропуски в столбце total_bedrooms средними значениями по категориям столбца ocean_proximity
```

```python
housing['total_bedrooms']= housing.groupby("ocean_proximity").transform(lambda x: x.fillna(x.mean())) 
```

```python
housing.isnull().sum()
```

3. Категорийные атрибуты:

```python
#а. Посчитать среднее кол-во комнат в строке (total_rooms/households)
```

```python
housing['avg_rooms_per_house']=housing['total_rooms']/housing['households']
```

```python
housing.head()
```

```python
housing['avg_rooms_per_house'] = housing['avg_rooms_per_house'] .apply(round)
housing.head()
```

```python
housing.groupby("avg_rooms_per_house").count().head(10)
```

```python
#б. Разбить полученный столбец на группы на ваше усмотрение (Например, 1 комната, 2 комнаты, 3 и более комнат)
```

```python
housing['rooms_in_house'] = pd.cut(housing['avg_rooms_per_house'], bins=[0,3,4,5,6,7,8,150], labels=['1,2,3 комнаты','4-комнаты','5-комнат','6-комнат', '7-комнат', '8-комнат','больше 9-комнат'])
```

```python
housing.head()
```

```python
housing.groupby("rooms_in_house").count()
```
