```python
#Есть отрицательные значения в количестве, видимо возвраты. Перед RFM-анализом и когортами уберем эти значения 
#из основного датасета
```

```python
Online_Retail_dt = Online_Retail_dt.query('Quantity > 0 and UnitPrice > 0')
```

```python
#проведем RFM-анализ (вместо М, будет G, так как Gross_revenue)
snapshot_date = Online_Retail_dt['InvoiceDate'].max() + dt.timedelta(days=1)
snapshot_date
```

```python
#Группиуем по покупателям, чтобы определить давность покупателя, частоту покупок, общее количество купленных единиц, общую выручку
Customer_groups=Online_Retail_dt.groupby(['CustomerID', 'InvoiceNo','InvoiceDate','Country'], as_index=False)\
.aggregate({'Quantity':'sum','Gross_revenue':'sum'})\
.groupby(['Country','CustomerID'], as_index=False)\
.aggregate({'InvoiceDate': lambda x: (snapshot_date - x.max()).days,'InvoiceNo':'count','Quantity':'sum','Gross_revenue':'sum'})\
.rename(columns={'InvoiceDate':'Recency','InvoiceNo':'Frequency'} )
Customer_groups.head()
```

```python
#Среднее количество товаров в покупке
Customer_groups['Quantity_per_order']=Customer_groups.Quantity/Customer_groups.Frequency
```

```python
#Средний чек
Customer_groups['Average_order_value']=Customer_groups.Gross_revenue/Customer_groups.Frequency
```

```python
Customer_groups.head()
```

```python
r_labels = range(3, 0, -1)
f_labels = range(1, 4)
g_labels = range(1, 4)
```

```python
r_groups = pd.qcut(Customer_groups['Recency'], q=3, labels=r_labels)
```

```python
f_groups = pd.qcut(Customer_groups['Frequency'], q=4, labels=f_labels, duplicates='drop')
```

```python
g_groups = pd.qcut(Customer_groups['Gross_revenue'], q=3, labels=g_labels, duplicates='drop')
```

```python
Customer_groups=Customer_groups.assign(R=r_groups, F=f_groups, G=g_groups)
Customer_groups.head()
```

```python
Customer_groups['Score'] = Customer_groups[['R', 'F', 'G']].sum(axis=1)
```

```python
Customer_groups['Segment'] = Customer_groups.apply(lambda x: str(x['R']) + str(x['F']) + str(x['G']), axis=1)
```

```python
Customer_groups.head()
```

```python
Customer_groups['Score'].hist()
```

```python
#Сегментация по итоговым баллам
def rfm_level(df):
    if df['Score'] >= 8:
        return 'Top'
    elif ((df['Score'] >= 5) and (df['Score'] < 8)):
        return 'Middle'
    else:
        return 'Low'
```

```python
Customer_groups['Level'] = Customer_groups.apply(rfm_level, axis=1)
```

```python
#Ппроставим день первой покупки
Online_Retail_dt['CohortDay']=Online_Retail_dt.groupby('CustomerID')['InvoiceDate'].transform('min')
Online_Retail_dt.tail()
```

```python
#Разница между первым днем покупки и последним днем покупки
Online_Retail_dt['CohortIndex'] = (Online_Retail_dt['InvoiceDate'] - Online_Retail_dt['CohortDay']).dt.days
Online_Retail_dt.tail()
```

```python
#Увеличим CohortIndex на 1, чтобы выровнять ситуацию, потому что есть покупатели, 
#у которых первый и последний день покупок совпадают
Online_Retail_dt['CohortIndex'] = Online_Retail_dt['CohortIndex'].add(1)
Online_Retail_dt.head()
```

```python
def get_month(x): return dt.datetime(x.year, x.month, 1)
```

```python
#Будем строить когорты по месяцам
Online_Retail_dt['InvoiceMonth'] = Online_Retail_dt['InvoiceDate'].apply(get_month)
```

```python
Online_Retail_dt['CohortMonth']  = Online_Retail_dt.groupby('CustomerID')['InvoiceMonth'].transform('min')
Online_Retail_dt.tail()
```

```python
#Переведем CohortIndex в месяцы вместо дней
def get_date_int(df, column):
    year = df[column].dt.year
    month = df[column].dt.month
    day = df[column].dt.day
    return year, month, day
invoice_year, invoice_month, _ = get_date_int(Online_Retail_dt, 'InvoiceMonth')
cohort_year, cohort_month, _ = get_date_int(Online_Retail_dt, 'CohortMonth')
years_diff = invoice_year - cohort_year
months_diff = invoice_month - cohort_month
```

```python
Online_Retail_dt['CohortIndex'] = years_diff * 12 + months_diff + 1
Online_Retail_dt.tail()
```

```python
Online_Retail_Cohort=Online_Retail_dt.groupby(['CohortMonth','CohortIndex'])['CustomerID'].nunique().reset_index()
Online_Retail_Cohort.head()
```

```python
#Удержание пользователей в каждой когорте
Online_Retail_Cohort_counts = Online_Retail_Cohort.pivot(index='CohortMonth', columns='CohortIndex', values='CustomerID')
Online_Retail_Cohort_counts
```

```python
#Покупатели, привлеченные в самый первый месяц - самые преданные. Но лучше проверить это в относительном выражении
```

```python
Online_Retail_Cohort_sizes = Online_Retail_Cohort_counts.iloc[:,0]
Online_Retail_Cohortretention = Online_Retail_Cohort_counts.div(Online_Retail_Cohort_sizes, axis=0)
```
