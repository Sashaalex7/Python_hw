```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
def rating(row):
    if row['rating'] <=2:
        return 'низкий рейтинг'
    if row['rating']<=4 and row['rating']>2:
        return 'средний рейтинг'
    if row['rating'] >4:
        return 'высокий рейтинг'
```

```python
ratings['rating_class']=ratings.apply(rating, axis=1)
```

```python
ratings.head()
```

```python
ratings['rating_class'].value_counts()
```

Задание 3

```python
ratings_filtered_userId=ratings.groupby('userId').count()
ratings_filtered_userId.head()
```

```python
ratings_filtered_userId=ratings_filtered_userId[ratings_filtered_userId['timestamp']>100]
ratings_filtered_userId.head()
```

```python
ratings_filtered_userId=ratings_filtered_userId.reset_index()
ratings_filtered_userId.head()
```

```python
ratings_filtered_userId['userId'].tolist()
```

```python
rating_movie_fan=ratings[ratings['userId'].isin(ratings_filtered_userId['userId'].tolist())]
rating_movie_fan.head()
```

```python
rating_movie_fan_timestamp=rating_movie_fan.groupby('userId').agg({'timestamp': ['min', 'max']})
rating_movie_fan_timestamp.head()
```

```python
import time
```

```python
def Lifetime(row):
    Lft= int(row['timestamp']['max'])-int(row['timestamp']['min'])
    return time.localtime(Lft).tm_mday
```

```python
rating_movie_fan_timestamp['lifetime']=rating_movie_fan_timestamp.apply(Lifetime, axis=1)
rating_movie_fan_timestamp
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
movies = pd.read_csv('movies.csv')
movies.drop_duplicates(subset = 'movieId', keep = 'first', inplace = True)
movies.head()
```

```python
full_rating=ratings.merge(movies, how = 'left', on = 'movieId')
```

```python
import numpy as np
```

```python
years=np.arange(1950, 2011).tolist()
```

```python
def production_year(row):
    for year in years:
        if str(year) in row['title']:
            return year
    return '1900'
```

```python
full_rating['year']=full_rating.apply(production_year, axis=1)
```

```python
full_rating.head()
```

```python
full_rating.groupby('year').mean()[['rating']].sort_values('rating', ascending=False).head()
```
